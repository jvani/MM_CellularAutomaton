```python
import os
import seaborn
import pandas as pd
import matplotlib.pyplot as pl
import scipy.constants as constants
from scipy.stats import ks_2samp
pd.options.display.max_rows = 10000
%matplotlib inline
```

```python
os.listdir('Data/')
```

```python
cols = ['Timestamp', 'Station', 'District', 'FreewayNum', 'Direction', 'LaneType',
        'StationLength', 'Samples', 'PercObs', 'TotalFlow', 'AvgOcc', 'AvgSpeed']
stations = [500010011, 500010021, 500011112, 500011121, 500011131, 500011143, 500011212, 500011222]
df = pd.read_csv('Data/d05_text_station_5min_2016_08_05.txt', usecols=range(12), names=cols)
df = df[df['Station'].isin(stations)]
```

```python
df.head(2)
```

```python
fig, ([ax1, ax2, ax3, ax4],
      [ax5, ax6, ax7, ax8]) = pl.subplots(nrows=2, ncols=4, figsize=(16, 8))
fig.tight_layout(w_pad=2, h_pad=4)

for ax, i in zip(fig.axes, df.Station.unique()):
    ax.scatter(df[df['Station'] == i].AvgOcc * 100, df[df['Station'] == i].TotalFlow, label=i, c='k', alpha=0.4)
    ax.set_title('Flow v. Occ. at: '+ str(i))
    ax.set_xlim(0, 50)
    ax.set_ylim(0, 120)
    ax.set_xlabel('Occupancy (%)')
    ax.set_ylabel('Flow (veh/5min)')
fig.delaxes(ax8)
```

```python
# Station 500011121: 1 lane, 1.515 miles, 70Mph limit, 

avg_speed = (df[df['Station'] == 500011121].AvgSpeed.mean() * constants.mile) / 1000 # Avg speed (kph).
road_length = 1.515 * constants.mile # Road length (m).

cell_len = 7.5 # Cell length (m).
iter_speed = 4.5 # Avg cell speed per iter.

cell_len * iter_speed * 3600 / 1000

cell_counts = road_length / cell_len
cell_counts
```

```python
data_output = pd.read_csv('Data/singlelane_varyinginput.csv',
                          names=['flow', 'occ', 'avgspeed'], usecols=[0,1,2])

data_output_slow = pd.read_csv('Data/singlelaneslow_varyinginput.csv',
                               names=['flow', 'occ', 'avgspeed'], usecols=[0,1,2])

data_output_slow1 = pd.read_csv('Data/singlelane_slow1.csv',
                                names=['flow', 'occ', 'avgspeed'], usecols=[0,1,2])
data_output_slow2 = pd.read_csv('Data/singlelange_slow2.csv',
                                names=['flow', 'occ', 'avgspeed'], usecols=[0,1,2])

output = pd.read_csv('Nagel-Schreckenberg-simulation-master/output.csv', 
                     names=['flow', 'occ', 'avgspeed'], usecols=[0,1,2])

output.head()
```

```python
fig, ([ax1, ax2],
      [ax3, ax4]) = pl.subplots(nrows=2, ncols=2, figsize=(16, 16))

ax1.scatter(data_output.occ, data_output.flow, c='b', alpha=0.6)
ax2.scatter(data_output_slow.occ, data_output_slow.flow, c='b', alpha=0.6)
ax3.scatter(data_output_slow1.occ, data_output_slow1.flow, c='b', alpha=0.6)
ax4.scatter(output.occ, output.flow, c='b', alpha=0.6)

for ax in fig.axes:
    ax.set_title('Flow v. Occ. at: '+ str(i))
    ax.set_xlim(0, 50)
    ax.set_ylim(0, 120)
    ax.set_xlabel('Occupancy (%)')
    ax.set_ylabel('Flow (veh/5min)')
    ax.scatter(df[df['Station'] == 500011121].AvgOcc * 100, 
            df[df['Station'] == 500011121].TotalFlow, label=i, c='k', alpha=0.4)
```

```python
fig, ([ax1, ax2],
      [ax3, ax4]) = pl.subplots(nrows=2, ncols=2, figsize=(16, 16))

ax1.scatter(data_output.avgspeed*7.5 * 3.6, data_output.flow, c='b', alpha=0.6)
ax2.scatter(data_output_slow.avgspeed*7.5 * 3.6, data_output_slow.flow, c='b', alpha=0.6)
ax3.scatter(data_output_slow1.avgspeed*7.5 * 3.6, data_output_slow1.flow, c='b', alpha=0.6)
ax4.scatter(output.avgspeed*7.5 * 3.6, output.flow, c='b', alpha=0.6)

for ax in fig.axes:
    ax.set_title('Flow v. Avg Speed. at: '+ str(i))
    ax.set_xlabel('Speed')
    ax.set_ylabel('Flow (veh/5min)')
    ax.scatter(df[df['Station'] == 500011121].AvgSpeed * 1.6, 
            df[df['Station'] == 500011121].TotalFlow, label=i, c='k', alpha=0.4)
```
